```python
import random, re


output = []
subject = []
artist = []
style = []
format = []
place = []

subject.extend(characters)

n=33

prompt_str  = 'SUBJECT, split screen with 3 views, front view, side view, back view, multi-split screen full body images with front, back a side views, photograph, hyperrealistic, photography, full body shot' 


for i in range(n):
    prompt = prompt_str + ' --v 5'  
    random_str = random.choice(subject)  
    prompt = re.sub(r'\bSUBJECT\b', random_str, prompt) 
    if len(artist) > 0:
      random_str = random.choice(artist)   
      prompt = re.sub(r'\bARTIST\b', random_str, prompt) 
    else:   
      prompt =  re.sub(r'ARTIST', ' ', prompt, flags=re.IGNORECASE)
    if len(format) > 0:
      random_str = random.choice(format)   
      prompt = re.sub(r'\bFORMAT\b', random_str, prompt) 
    else:   
      prompt = re.sub(r'\bFORMAT\b', '', prompt)
    if len(style) > 0:
      random_str = random.choice(style)   
      prompt = re.sub(r'\bSTYLE\b', random_str, prompt) 
    else:   
      prompt = re.sub(r'\bSTYLE\b', '', prompt)
    if len(place) > 0:
      random_str = random.choice(place)   
      prompt = re.sub(r'\bPLACE\b', random_str, prompt) 
    else:   
      prompt = re.sub(r'\bPLACE\b', '', prompt)   
    prompt = re.sub(r'[ ]+', ' ', prompt)     
    output.append(prompt)

for i in output:
  print(i)    
```
